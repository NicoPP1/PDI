# Introducción
En el presente trabajo se desarrolla un código para abrir una imagen, modificar la misma a nivel de pixel y mostrar la misma.

```python
#Librerias
import numpy as np
import imageio
import matplotlib.pyplot as plt

#importo imagen
img = imageio.imread('lenna.jpg')

#muestro imagen
print("Imagen original")
plt.imshow(img)
```

```python
print("Imagen modificada")
#img.shape -> (width, height, channels)
for i in range(img.shape[0]):    # for every col:
    for j in range(img.shape[1]):    # For every row
        img[i,j] = (i, j, 0) # set the colour accordingly

imageio.imsave("modificada.jpg",img)
plt.imshow(img)
```
